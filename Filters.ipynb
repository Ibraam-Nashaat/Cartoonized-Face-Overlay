## landmark code

```python
import cv2
import dlib
import numpy as np

# Load the cascade
face_cascade = cv2.CascadeClassifier("haarcascade_frontalface_default (1).xml")

# Load the shape predictor model
shape_predictor = dlib.shape_predictor("shape_predictor_68_face_landmarks.dat")

# Read the input image
image = cv2.imread("photo1.png")

# Convert the image to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Detect faces
faces = face_cascade.detectMultiScale(gray, 1.3, 4)

# Iterate over detected faces
for (x, y, w, h) in faces:
    # Convert the rectangle coordinates to dlib rectangle
    rect = dlib.rectangle(int(x), int(y), int(x + w), int(y + h))
    
    # Use the shape predictor to predict face landmarks
    landmarks = shape_predictor(gray, rect)
    
    # Convert the landmarks to a NumPy array
    for n in range(0, 68):
        x = landmarks.part(n).x
        y = landmarks.part(n).y
        cv2.circle(image, (x, y), 3, (0, 255, 0), -1)

    
    # Display the aligned face
    cv2.imshow("Aligned Face", image)
    cv2.waitKey(0)

cv2.destroyAllWindows()
```

## code for alpha blending


```python


import cv2 

image1 = cv2.imread('photo1.png')
image2 = cv2.imread('aa.PNG')

image1 = cv2.resize(image1, (image2.shape[1], image2.shape[0]))

# Set the alpha blending factor (0.0 for image1, 1.0 for image2)
alpha = 0.3

# Perform alpha blending
blended_image = cv2.addWeighted(image1, 0.5, image2, 1 - alpha, 0)

# Display the blended image
cv2.imshow('Blended Image', blended_image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
def Alpha_blending(background,upper):
    upper=cv2.resize(upper,(128,128))
    background=cv2.resize(background,(128,128))
    upper = upper.astype(float)
    _, alpha = cv2.threshold(upper, 100, 255, cv2.THRESH_BINARY)
    background = background.astype(float)
    alpha = alpha.astype(float)
 
    # Multiply the foreground with the alpha matte
    upper = cv2.multiply(alpha, upper)
 
    # Multiply the background with ( 1 - alpha )
    background = cv2.multiply(1.0 - alpha, background)
 
    # Add the masked foreground and background.
    outImage = cv2.add(upper, background)
 
    # Display image
    plt.imshow("outImg", outImage/255)
    
```

```python
# Filters

# A helper function to resize an image to a given size
def resize_image(image, size):
    return cv2.resize(image, size)

# A generic function to apply a filter image to a face image at a given position and size
# The filter image is resized and adjusted according to the x_offset and y_offset parameters
def apply_filter(filter_img, image, x, y, w, h, x_offset, y_offset):
    w_face = w # the width of the face
    h_face = h # the height of the face
    w_filter = w_face + 1 # the width of the filter
    h_filter = int(0.50 * h_face) + 1 # the height of the filter
    filter_img = resize_image(filter_img, (w_filter, h_filter)) 
    for i in range(h_filter):
        for j in range(w_filter):
            for k in range(3):
                if filter_img[i][j][k] < 235: # check if the pixel is not transparent
                    # adjust the position according to the offsets
                    image[y + i - int(y_offset * h_face)][x + j - int(x_offset * w_face)][k] = filter_img[i][j][k]
    return image 

# A specific function to put a hat on a face image
def put_hat(hat, image, x, y, w, h):
    # call the apply_filter function with no offsets
    return apply_filter(hat, image, x, y, w, h, 0, 0.4)

# A specific function to put a glass on a face image
def put_glass(glass, image, x, y, w, h):
    # call the apply_filter function with a negative y_offset
    return apply_filter(glass, image, x, y, w, h, 0, -0.2)

# A specific function to put a cartoon glass on a face image
def put_cartoon_glass(glass, image, x, y, w, h):
    w_face = w # the width of the face
    h_face = h # the height of the face
    w_glass = int(0.9 * w_face) + 1 # the width of the glass
    h_glass = int(0.20 * h_face) + 1 # the height of the glass
    glass = resize_image(glass, (w_glass, h_glass))
    for i in range(h_glass):
        for j in range(w_glass):
            for k in range(3):
                if glass[i][j][k] < 235: # check if the pixel is not transparent
                    # adjust the position according to a negative y_offset and a small negative x_offset
                    image[y + i - int(-0.30 * h_face)][x + j-int(-0.05*w_face)][k] = glass[i][j][k]
    return image 

# A specific function to put a mask on a face image
def put_mask(mask, image, x, y, w, h):
    w_face = w 
    h_face = h 
    w_mask = w_face + 1 
    h_mask = int(.7* h_face) + 1 
    mask = resize_image(mask, (w_mask, h_mask)) 
    for i in range(h_mask):
        for j in range(w_mask):
            for k in range(3):
                if mask[i][j][k] < 235: 
                    # adjust the position according to a negative y_offset
                    image[y + i - int(-0.40 * h_face)][x + j][k] = mask[i][j][k]
    return image 
```
